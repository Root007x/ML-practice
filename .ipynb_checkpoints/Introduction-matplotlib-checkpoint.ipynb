## Matplolib example workflow

```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4]
y = [11,22,33,44]

# oop style
fig, ax = plt.subplots(figsize=(10,10)) # height and width
ax.plot(x,y)
ax.set(title="Simple Plot",
      xlabel="x-axis",
      ylabel="y-axis")

fig.savefig("figure/sample-plot.png")
```

## Making figures with Numpy arrays
* Line plot
* Scatter Plot
* Bar plot
* Histogram
* subplot

```python
import numpy as np
```

```python
x = np.linspace(0,10,100)
x
```

```python
# plot the data and create a line plot
fig, ax = plt.subplots()
ax.plot(x,x**2)
```

```python
# use the same data to make scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x))
```

```python
# Another scatter plot
fig, ax = plt.subplots()
ax.scatter(x,np.sin(x))
```

```python
# Make a plot from dictionary
nut_butter = {
    "A butter" : 10,
    "B butter" : 20,
    "C butter" : 30
}
fig, ax = plt.subplots()
ax.bar(nut_butter.keys(),nut_butter.values()) #(x,y)
ax.set(title="Lopu shop",
      ylabel="Price")
plt.show()
```

```python
fig, ax = plt.subplots()
ax.barh(list(nut_butter.keys()),list(nut_butter.values()))
```

```python
# Histogram

x = np.random.randn(1000)
```

```python
fig, ax = plt.subplots()
ax.hist(x)
plt.show()
```

## Two option for subplots

```python
# subplot option 1
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2,ncols=2,figsize=(10,5))

# plot to each different axis
ax1.plot(x,x/2)
ax2.scatter(np.random.random(10),np.random.random(10))
ax3.bar(nut_butter.keys(),nut_butter.values())
ax4.hist(x)
plt.show()
```

```python
# subplots option 2
fig, ax = plt.subplots(nrows=2,ncols=2,figsize=(10,5))

# plot to each different index
ax[0,0].plot(x,x/2)
ax[0,1].scatter(np.random.random(10),np.random.random(10))
ax[1,0].bar(nut_butter.keys(),nut_butter.values())
ax[1,1].hist(x)
plt.show()
```

## Plotting from pandas DataFrames

```python
import pandas as pd
```

```python
df = pd.read_csv("datasheet/car-sales.csv")
df
```

```python
df["Price"] = df["Price"].str.replace('[\$\,\.]','',regex=True)
df
```

```python
# Remove last tow zeros
df["Price"] = df["Price"].str[:-2]
df
```

```python
df["Sale Date"] = pd.date_range("1/1/2020",periods=len(df))
df
```

```python
df["Total Sales"] = df["Price"].astype(int).cumsum() # cumalative sum
df
```

```python
df.plot(x="Sale Date",y="Total Sales")
```

```python
df['Price'] = df['Price'].astype(int)
df.plot(x="Odometer (KM)",y="Price",kind="scatter")
```
